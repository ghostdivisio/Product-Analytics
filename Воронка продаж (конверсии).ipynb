Что касается воронки, отражающей этапы по неделям, то больше всего человек посетило сайт в первую рабочую неделю октября 2022 года и в ту же неделю произошло больше всего регистраций. Однако больше всего регистраций на игры было в 3 неделю 2023 года (16-22 января 2023). Надо отметить, что немало заявок, а именно 33, было подано тогда же, когда больше всего пользователей посещало сайт и регистрировалось на нём. В некотором плане, можно выделить начало октября как точку роста, связанную с сезонностью и началом полноценной осени. Больше всего игр (если смотреть по неделям) состоялось в январе 2023, это новогодние каникулы и конец января. Отсюда можно сделать вывод, что игры лучше всего проводить в новогодние каникулы и близкое к этому времени периоды, поскольку у пользователей появляется больше свободного времени и улучшается настроение.

Исходя из этого можно выделить следующие точки роста продукта: первое и основное-это улучшить сайт, например, сделать более простую регистрацию. Важно что-то менять на этом этапе потому, что уже здесь мы видим самый большой отсев 75%. Далее можно упростить способ подачи заявки на игру, чтобы достичь роста на этапе "Подача заявки на игру". И наконец не менее значимая точка роста-заставить зарегистрировавшегося пользователя поучаствовать в игре. Здесь можно либо увеличить количество игр, сделав выбор более разнообразным, либо не менять количество, а сосредоточиться на их качестве и привлекательности для пользователя. Таким образом, можно выделить три основных точки роста продукта от самого главного к менее значительному: улучшение сайта (первый этап), увеличить количественно/качественно игры, упростить способ подачи заявок на игры. Касательно сезонности, больше всего людей посетило сайт в январе 2023, рекорд по регистрациям на сайте - начало октября 2022, а поучаствовало в играх - в январе 2023 года.
